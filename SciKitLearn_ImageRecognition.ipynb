### Support Vector Machines for Image Classification

Image instances in a dataset are points in a multidimensional space. So the model built by a classifier can be a surface (hyperplane) that separates instances (points) of one class from the rest.

`Support Vector Machines (SVMs)` are supervised learning methods that obtain these hyperplanes in an optimal way. 

Here we will apply SMV to image recognition, a classic problem with a very large dimensional space (the value of each image pixel is a feature). Our learning set will be a group of labeled images of peoples' faces, and we will try to learn a model that can predict the label of unseen instances. Pixel values will be our learning attribute and the individual's label will be our target class.

```python
import sklearn as sk
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import fetch_olivetti_faces
faces = fetch_olivetti_faces()
```

```python
print faces.DESCR
```

Let's take a look at the data set components and shapes:

```python
print faces.keys()
print faces.images.shape
print faces.data.shape
print faces.target.shape
```

Let's look at whether the data (pixel values) need to be scaled/normalized prior to learning. It does not since values range from 0-1

```python
print np.max(faces.data)
print np.min(faces.data)
print np.mean(faces.data)
```

Let's plot some of the faces:

```python
def print_faces(images, target, top_n):
    # set up figure size in inches
    fig = plt.figure(figsize=(12,12))
    fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05)
    for i in range(top_n):
        # plot the images in a matrix of 20x20
        p = fig.add_subplot(20, 20, i+1, xticks=[], yticks=[])
        p.imshow(images[i], cmap=plt.cm.bone)
        # label the image with the target value
        p.text(0, 14, str(target[i]))
        p.text(0, 60, str(i))
```

```python
print_faces(faces.images, faces.target, 40)
```

The `SVC` implementation (of an SVM) has various parameters. The most relevant is `kernel`, which defines the classifier's kernel (think of the kernel functions as different similarity measures between instances). The default `SVC` kernel is the `rbf` kernel, which allows us to model nonlinear problems. We will start with the simplest one: `linear`

```python
from sklearn.cross_validation import train_test_split
from sklearn.svm import SVC
svc_1 = SVC(kernel='linear')
```

```python
X_train,X_test, y_train, y_test = train_test_split(faces.data, faces.target, test_size = 0.25, random_state=0)
```

```python
print X_train.shape
print X_test.shape
print y_train.shape
print y_test.shape
```

Let's see how `train_test_split` did in randomly selecting from the set of `0 - 39` images to train on (image index, count):

```python
y = np.bincount(y_train)
ii = np.nonzero(y)[0]      # gets indices of y
print zip(ii,y[ii]) 
```

Define a function to evaluate K-fold cross-validation:

```python
from sklearn.cross_validation import cross_val_score, KFold
from scipy.stats import sem
```

```python
def evaluate_cross_validation(clf, X, y, K):
    #create a k-fold cross validation iterator
    cv = KFold(len(y), K, shuffle=True, random_state=0)
    # by default, the score used is the one returned by score method of estimator (`accuracy`)
    scores = cross_val_score(clf, X, y, cv=cv)
    print scores
    print("Mean score: {0:.3f} (+/- {1:.3f})").format(np.mean(scores),sem(scores))
    
```

```python
evaluate_cross_validation(svc_1, X_train, y_train, 5)
```

Cross-valiation with 5 folds gets good results (accuracy of 0.933). The above few steps gets us a face-classifier

We will also define a function to perform training on the training set and evaluate the performance on the testing set:

```python
from sklearn import metrics
```

```python
def train_and_evaluate(clf, X_train, X_test, y_train, y_test):
    clf.fit(X_train,y_train)
    print "Accuracy on training set:"
    print clf.score(X_train, y_train)
    print "Accuracy on testing set:"
    print clf.score(X_test, y_test)
    
    y_pred = clf.predict(X_test)
    
    print "Classification Report:"
    print metrics.classification_report(y_test, y_pred)
    print "Confusion Matrix"
    print metrics.confusion_matrix(y_test, y_pred)
```

```python
train_and_evaluate(svc_1, X_train, X_test, y_train, y_test)
```

Let's try to classify the faces as people with and without glasses.

```python
glasses = [(10,19), (30,32), (37,38), (50,59), (63,64), (69,69), (120,121), (124,129), (130,139), (160,161),
           (164,169), (180,182), (185,185), (189,189), (190,192), (194,194), (196,199), (260,269), (270,279),
           (300,309), (330, 339), (358, 359), (360,369)]
```

From the `glasses` list, create a new target array that makes with `1` the faces with glasses and 0 for the faces without glasses:

```python
def create_target(segments):
    # create new y array of target size initialized with zeros
    y = np.zeros(faces.target.shape[0])
    # insert a 1 for those images with glasses
    for (start, end) in segments:
        y[start:end + 1] = 1
    return y
```

```python
target_glasses = create_target(glasses)
```

So we must perform the training-testing split again:

```python
X_train, X_test, y_train, y_test = train_test_split(faces.data, target_glasses, test_size=0.25, random_state=0)
```

```python
print X_train.shape
print X_test.shape
print y_train.shape
print y_test.shape
```

Let's create a new SVC classifier, and train it with the new target vector:

```python
svc_2 = SVC(kernel='linear')
```

Check the performance with cross-validation on our training sets by the following:

```python
evaluate_cross_validation(svc_2, X_train, y_train, 5)
```

```python
train_and_evaluate(svc_2, X_train, X_test, y_train, y_test)
```

Results from classification report has 67 without glasses, 33 with glasses. Confusion matrix shows that we mis-classified a person without glasses as wearing glasses (false positive). 

Let's separate all the images of the same person, wearing glasses and not. We'll separate all the images of the same person. We will also separate all the images of the same person, the ones with indixes from 30 to 39, train by using the remaining instances, and evaluate on our new 10 instances set. With this experiment we will try to discard the fact that it is remembering faces, not glassed-related features.

```python
X_test = faces.data[30:40]
y_test = target_glasses[30:40]
```

```python
print y_test.shape[0]
```

Now remove the person 30-39 from the training set

```python
select = np.ones(target_glasses.shape[0])
select[30:40] = 0
```

```python
X_train = faces.data[select == 1]
y_train = target_glasses[select == 1]
```

```python
print X_train.shape
print y_train.shape
```

```python
svc_3 = SVC(kernel='linear')
```

```python
train_and_evaluate(svc_3, X_train, X_test, y_train, y_test)
```

According to results above, we mis-classified an image in our test set that was wearing glasses as not wearing glasses. Let's investigate why by evaluating the images:

```python
eval_faces = [np.reshape(a, (64,64)) for a in X_test]
y_pred = svc_3.predict(X_test)
```

```python
print_faces(eval_faces, y_pred, 10)
```

So, image 8 has glasses and was classified as having no glasses. In reviewing the image, the glasses don't look as defined as the others.

Let's try using the `rbf` kernel in the SVM classifier instead and check the results:

```python
svc_4 = SVC(kernel='rbf')
```

```python
X_train, X_test, y_train, y_test = train_test_split(faces.data, target_glasses, test_size=0.25, random_state=0)
```

```python
print X_train.shape
print X_test.shape
print y_train.shape
print y_test.shape
```

```python
evaluate_cross_validation(svc_4, X_train, y_train, 5)
```

```python
train_and_evaluate(svc_4, X_train, X_test, y_train, y_test)
```

As can be seen above, the `rbf` classified all those wearing glasses as not wearing glasses. A bad classifier for this situation.
